```python
import scipy.stats as stats
z = np.abs(stats.zscore(data))
print(z)
print(z.max())
print(z.min())
#print(np.where(z>3))
data_f=data[(z<3).all(axis=1)]
data_f
```

```python
phw=0.11
tempw=0.10
turbw=0.08
dow=0.07
nw=0.10
fcw=0.16
tw=phw+tempw+turbw+dow+nw+fcw
#print(tw)
phv=(data_f['PH']*phw)/tw
tempv=(data_f['TEMP']*tempw)/tw
turbv=(data_f['TURBIDITY']*turbw)/tw
dov=(data_f['DISSOLVEDOXYGEN']*dow)/tw
nv=(data_f['NITRATES']*nw)/tw
fcv=(data_f['COLIFORM']*fcw)/tw
wqi=phv+tempv+turbv+dov+nv+fcv
print(wqi)
```

```python
data_f['WQI']=wqi
data_f
```

```python
data_f['WQC']=data_f.WQI.apply(lambda x:('VB' if (25>=x>=0)  
                                 else('B' if  (50>=x>25) 
                                      else('M' if (70>=x>50)
                                          else('G' if (90>=x>70)
                                               else('EX' if (100>=x>90) 
                                                    else 'NO'))))))                                          
```

```python
import pandas as pd
import numpy as np
dataset=data_f
y=dataset.iloc[:,-2:-1].values
x=dataset.iloc[:,-8:-2].values
print(x.shape)
print(y.shape)
#print(y)
#print(x)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=0) 
#print(X_test.shape)
#print(y_test.shape)
from sklearn.linear_model import LinearRegression 
regr = LinearRegression()
regr.fit(X_train, y_train)
pred=regr.predict(X_test)
pred
```

```python
np.savetxt("predict_load.csv", pred, delimiter=",")
```

```python
print('Train Score: ', regr.score(X_train, y_train))  
print('Test Score: ', regr.score(X_test, y_test))  
```

```python
from sklearn import metrics
print('Intercept:', regr.intercept_)
print('Coefficients:', regr.coef_)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
